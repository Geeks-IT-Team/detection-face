```python
from ultralytics import YOLO
from facenet_pytorch import InceptionResnetV1
from PIL import Image as PILImage
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import torch
import numpy as np
import cv2
import pickle
import os
from IPython.display import display
from PIL import Image
from collections import defaultdict


# Chargement du modèle préentrainé vggface2 de FACENET
facenet = InceptionResnetV1(pretrained='vggface2').eval()

# Chargement du modèle de yolo pour la detection de visage
yolo    = YOLO("yolov8n-face.pt")

# Chargement du modèle de Mediapipe pour l'alignement de visage
face_landmarker = vision.FaceLandmarker.create_from_options(
    vision.FaceLandmarkerOptions(
        base_options=python.BaseOptions(model_asset_path="face_landmarker.task"),
        running_mode=vision.RunningMode.IMAGE
    )
)

print("Modèles chargés !")


# DETECTION
def detecter_visage(image_rgb):
    h, w = image_rgb.shape[:2]
    results  = yolo(image_rgb, verbose=False)[0]
    upscaled = False

    if len(results.boxes) == 0:
        image_up = cv2.resize(image_rgb, (w * 2, h * 2))
        results  = yolo(image_up, verbose=False)[0]
        upscaled = True

    if len(results.boxes) == 0:
        return None

    best = max(results.boxes, key=lambda b: float(b.conf[0]))
    x1, y1, x2, y2 = map(int, best.xyxy[0])

    if upscaled:
        x1, y1, x2, y2 = x1//2, y1//2, x2//2, y2//2

    return (x1, y1, x2 - x1, y2 - y1)


# ALIGNEMENT
def aligner_visage(image_rgb, bbox):
    x, y, w, h = bbox
    marge = int(max(w, h) * 0.25)
    x1 = max(0, x - marge);  y1 = max(0, y - marge)
    x2 = min(image_rgb.shape[1], x + w + marge)
    y2 = min(image_rgb.shape[0], y + h + marge)
    crop = image_rgb[y1:y2, x1:x2]

    if crop.size == 0:
        return image_rgb, (0, 0, image_rgb.shape[1], image_rgb.shape[0])

    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=crop)
    result   = face_landmarker.detect(mp_image)

    if not result.face_landmarks:
        return crop, (x1, y1, x2, y2)

    landmarks = result.face_landmarks[0]
    h_crop, w_crop, _ = crop.shape
    lx = int(landmarks[33].x  * w_crop);  ly = int(landmarks[33].y  * h_crop)
    rx = int(landmarks[263].x * w_crop);  ry = int(landmarks[263].y * h_crop)

    angle   = np.degrees(np.arctan2(ry - ly, rx - lx))
    centre  = (float((lx + rx) / 2), float((ly + ry) / 2))
    M       = cv2.getRotationMatrix2D(centre, angle, 1.0)
    aligned = cv2.warpAffine(crop, M, (w_crop, h_crop))

    return aligned, (x1, y1, x2, y2)


# EMBEDDING
def extraire_embedding(crop_rgb):
    if crop_rgb is None or crop_rgb.size == 0:
        return None
    img        = PILImage.fromarray(crop_rgb).resize((160, 160))
    img_tensor = torch.tensor(np.array(img)).permute(2, 0, 1).float() / 255.0
    with torch.no_grad():
        embedding = facenet(img_tensor.unsqueeze(0))
    return embedding[0].numpy()


# CONSTRUCTION BASE MULTI-EMBEDDINGS
base_de_visages = defaultdict(list)
dossier_photos  = "photos_con"

print("\nConstruction de la base multi-embeddings...\n")

for fichier in sorted(os.listdir(dossier_photos)):
    if not fichier.lower().endswith(('.png', '.jpg', '.jpeg')):
        continue

    # Extraire le nom : "dupont_1.jpg" → "dupont"
    stem = os.path.splitext(fichier)[0]          # "dupont_1"
    nom  = stem.rsplit('_', 1)[0] if '_' in stem else stem  # "dupont"

    chemin   = os.path.join(dossier_photos, fichier)
    image_cv = cv2.imread(chemin)
    if image_cv is None:
        print(f"[ERREUR]  Impossible de lire : {fichier}")
        continue

    image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)
    bbox      = detecter_visage(image_rgb)

    if bbox is None:
        print(f"[MANQUÉ]  Aucun visage : {fichier}")
        continue

    crop, (x1, y1, x2, y2) = aligner_visage(image_rgb, bbox)
    embedding = extraire_embedding(crop)

    if embedding is None:
        print(f"[MANQUÉ]  Embedding échoué : {fichier}")
        continue

    base_de_visages[nom].append(embedding)

    # Affichage
    img_aff = image_cv.copy()
    cv2.rectangle(img_aff, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img_aff, f"{nom} #{len(base_de_visages[nom])}",
                (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
    display(Image.fromarray(cv2.cvtColor(img_aff, cv2.COLOR_BGR2RGB)))
    print(f"[OK]  {nom} — photo #{len(base_de_visages[nom])} ({fichier})")

# Résumé
print("\n── Résumé ──────────────────────────")
for nom, embeddings in base_de_visages.items():
    print(f"  {nom} : {len(embeddings)} embedding(s)")
print(f"\nTotal : {len(base_de_visages)} personne(s)")

with open("base_visages_multiple_face.pkl", "wb") as f:
    pickle.dump(dict(base_de_visages), f)
print("Base sauvegardée !")
```

```python
from ultralytics import YOLO
from facenet_pytorch import InceptionResnetV1
from PIL import Image as PILImage
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import torch
import numpy as np
import cv2
import pickle
from IPython.display import display
from PIL import Image


# CHARGEMENT BASE + MODÈLES
with open("base_visages_multiple_face.pkl", "rb") as f:
    base_de_visages = pickle.load(f)
print(f"Base chargée : {list(base_de_visages.keys())}")

yolo    = YOLO("yolov8n-face.pt")
facenet = InceptionResnetV1(pretrained='vggface2').eval()

face_landmarker = vision.FaceLandmarker.create_from_options(
    vision.FaceLandmarkerOptions(
        base_options=python.BaseOptions(model_asset_path="face_landmarker.task"),
        running_mode=vision.RunningMode.IMAGE
    )
)

print("YOLOv8-face + FaceNet + Landmarker chargés !")


# ALIGNEMENT  (MediaPipe Landmarker)
def aligner_visage(image_rgb, bbox):
    x, y, w, h = bbox

    marge = int(max(w, h) * 0.25)
    x1 = max(0, x - marge)
    y1 = max(0, y - marge)
    x2 = min(image_rgb.shape[1], x + w + marge)
    y2 = min(image_rgb.shape[0], y + h + marge)

    crop = image_rgb[y1:y2, x1:x2]
    if crop.size == 0:
        return image_rgb, (0, 0, image_rgb.shape[1], image_rgb.shape[0])

    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=crop)
    result   = face_landmarker.detect(mp_image)

    if not result.face_landmarks:
        return crop, (x1, y1, x2, y2)

    landmarks = result.face_landmarks[0]
    h_crop, w_crop, _ = crop.shape

    left_eye  = landmarks[33]
    right_eye = landmarks[263]
    lx = int(left_eye.x  * w_crop);  ly = int(left_eye.y  * h_crop)
    rx = int(right_eye.x * w_crop);  ry = int(right_eye.y * h_crop)

    angle  = np.degrees(np.arctan2(ry - ly, rx - lx))
    centre = (float((lx + rx) / 2), float((ly + ry) / 2))

    M       = cv2.getRotationMatrix2D(centre, angle, 1.0)
    aligned = cv2.warpAffine(crop, M, (w_crop, h_crop))

    return aligned, (x1, y1, x2, y2)


# EMBEDDING
def extraire_embedding(crop_rgb):
    if crop_rgb is None or crop_rgb.size == 0:
        return None
    img        = PILImage.fromarray(crop_rgb).resize((160, 160))
    img_tensor = torch.tensor(np.array(img)).permute(2, 0, 1).float() / 255.0
    img_tensor = img_tensor.unsqueeze(0)
    with torch.no_grad():
        embedding = facenet(img_tensor)
    return embedding[0].numpy()


# DISTANCE COSINUS
def distance_cosinus(vec1, vec2):
    dot   = np.dot(vec1, vec2)
    norme = np.linalg.norm(vec1) * np.linalg.norm(vec2)
    return 1 - (dot / norme) if norme != 0 else 1.0


# IDENTIFICATION
def identifier_visage(crop_rgb, base, seuil=0.3):
    embedding_inconnu = extraire_embedding(crop_rgb)
    if embedding_inconnu is None:
        return "Erreur", 0.0

    meilleur_nom       = "Inconnu"
    meilleure_distance = seuil
    toutes_distances   = {}

    for nom, embeddings in base.items():
        # ── Compatibilité ancien format (un seul array) → convertir en liste
        if isinstance(embeddings, np.ndarray):
            embeddings = [embeddings]

        # ── Comparer avec chaque embedding
        distances = [distance_cosinus(embedding_inconnu, emb) for emb in embeddings]
        dist_min  = float(min(distances))  # ← float() évite l'ambiguïté numpy
        toutes_distances[nom] = round(dist_min, 3)

        if dist_min < meilleure_distance:
            meilleure_distance = dist_min
            meilleur_nom       = nom

    print(f"    Distances min : {toutes_distances}")
    confiance = max(0, (1 - meilleure_distance) * 100)
    return meilleur_nom, confiance


# RECONNAISSANCE SUR L'IMAGE
image_cv = cv2.imread('j.jpg')

# Redimensionner si nécessaire
h_img, w_img = image_cv.shape[:2]
max_dim = 1280
if max(w_img, h_img) > max_dim:
    scale    = max_dim / max(w_img, h_img)
    image_cv = cv2.resize(image_cv, (int(w_img * scale), int(h_img * scale)))

image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)

# ── Détection YOLOv8
results = yolo(image_rgb, verbose=False)[0]
faces   = results.boxes
print(f"\n{len(faces)} visage(s) détecté(s)")

for i, box in enumerate(faces):
    conf_detection = float(box.conf[0])

    # Ignorer les détections peu fiables
    if conf_detection < 0.4:
        continue

    x1, y1, x2, y2 = map(int, box.xyxy[0])
    bbox = (x1, y1, x2 - x1, y2 - y1)

    print(f"\nVisage {i+1} (confiance détection : {conf_detection:.2f}) :")
    visage_num = i+1
    espace = " "

    # ── Aligner
    crop, (ax1, ay1, ax2, ay2) = aligner_visage(image_rgb, bbox)
    if crop.size == 0:
        continue

    # ── Identifier
    nom, confiance_id = identifier_visage(crop, base_de_visages)
    print(f"  Résultat : {nom} ({confiance_id:.1f}%)")

    # ── Couleur : vert si reconnu, rouge si inconnu
    couleur = (0, 255, 0) if nom != "Inconnu" else (0, 0, 255)

    # ── Rectangle autour du visage (crop avec marge)
    cv2.rectangle(image_cv, (ax1, ay1), (ax2, ay2), couleur, 2)

    # ── Label avec fond coloré
    label = f"V{visage_num}{espace}{nom} {confiance_id:.0f}%" if nom != "Inconnu" else f"V{visage_num}{espace}{nom}"
    #label = f"{nom} {confiance_id:.0f}%"
    (tw, th), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.7, 2)
    cv2.rectangle(image_cv, (ax1, ay1 - th - 12), (ax1 + tw + 8, ay1), couleur, -1)
    cv2.putText(image_cv, label, (ax1 + 4, ay1 - 6),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

# ── Affichage
display(Image.fromarray(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)))
```
